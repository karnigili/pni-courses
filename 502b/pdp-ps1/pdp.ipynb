## Another XOR

In this section, we'll try to analyze the XOR solution for a different neural network architecture, briefly mentioned in class. In this network, the inputs both project to a single hidden unit. Then the inputs and the hidden unit (3 values in total) project to the output unit. We'll use logistic activations for all units. Though not depicted below, every hidden unit also has a bias term.

<img src="resources/net-xor-2.svg" />

Here are the equations for the network. Recall from PDP-02 that $\sigma$ is the logistic function. We index into vectors with a superscript, so the weight from input $i$ to the hidden layer is $w_1^i$.

\begin{align}
h &= \sigma(w_1 x + b_1) \\
\hat{y} &= \sigma(w_2 [x, h]' + b_2) \\
&= \sigma(w_2^1 x^1 + w_2^2 x^2 + w_2^3 h + b_2) \\
\end{align}

## First steps

We'll start by training a network and looking at hidden unit representations. Things have changed a little from the demo. We will use `utils.fit_model` to fit our model. It returns an array with a history of the versions of the model at every time step. Our loss function trains to a softer target of 0.05 / 0.95. Finally, our network adds random noise to the hidden units during training.

```python
torch.manual_seed(42)
model = utils.Rumelhart(nouns, relations, ylabels)
def loss():
    # Per Semantic Cognition, we have a softer target of 0.05 / 0.95 instead of 0 / 1.
    Yloss = Y * 0.9 + 0.05
    se = (model(X) - Yloss) ** 2
    return se.sum() / X.shape[0]
model_history = utils.fit_model(loss, model, lr=4., num_epochs=4000)
```

## Changing word frequency

Our next focus in this problem set is to explore changes in representations due to changes in the training set. We'll focus on changing the frequency of words for now. A criticism of the Rumelhart & Todd 1993 model was that it learned general words (like **animal** and **plant**) before basic-level or specific words (**tree** or **canary**), at odds with the developmental literature. An approach to address this was to make sure that different words all had the same frequency. Without an adjustment to this frequency, we would otherwise train 8x on the word `animal` vs on `sparrow`, an imbalance that might be driving the learning of general-level words.

So, we'll make `isa-general`, `isa-basic`, and `isa-specific` all happen with equal frequency. We accomplish this by making `isa-general` 1/8th as likely and `isa-basic` 1/4th as likely. We're also making words less likely, mirroring the frequencies specified in "Semantic Cognition".

```python
nouns, relations, xlabels, ylabels, X, Y, frequency = utils.load_semantic_cognition_data(equal_word_frequency=True)

for ridx, relation in enumerate(relations):
    percent = frequency[ridx] / sum(frequency[:len(relations)])
    print(f'{relation} {percent.item()*100:.2f}%')
```

Now we'll train a network on this dataset. Here's some sample code. We'll change our loss function so instead of simply taking a mean over our samples, we'll use the frequency to determine the contribution of samples to the training process. While this is different from sampling with the above frequency, it's still an appropriate way to take the expectation of the error.

```python
torch.manual_seed(42)

model = utils.Rumelhart(nouns, relations, ylabels)

def loss():
    # Per Semantic Cognition, we have a softer target of 0.05 / 0.95 instead of 0 / 1.
    Yloss = Y * 0.9 + 0.05

    se = (model(X) - Yloss) ** 2

    # Frequency modulates our sum, no longer just a mean.
    return (frequency[:, None] * se).sum()

model_history = utils.fit_model(loss, model, lr=4., num_epochs=4000)
```

Here's example code plotting the activation for `sparrow can fly` over the course of training. Note that you can use the `predict_query(noun, relation)` function on any model. It returns predictions for the noun and relation argument supplied. To index into this array of predictions, we use the Python function `ylabels.index()` which returns the index of the argument in the list `ylabels`.

```python
activation = []
for m in model_history:
    yhat = m.predict_query('sparrow', 'can')
    activation.append(yhat[ylabels.index('fly')])

plt.plot(activation, label='sparrow can fly')
plt.xlabel('# epochs')
plt.ylabel('activation')
plt.legend();
```
