```python
def plot_mat(genotype, raw, times, sines, pulses, xlim=None, ylim=None):
    # from toturial
    fig,axs = plt.subplots(sharex=True, figsize=(10,2))

    # plot raw mic trace
    axs.plot(times,raw[:,0], color='grey') 
    axs.set_ylabel('Amplitude (a.u.)')
    axs.set_xlabel('Time (sec)')

    #unpack sine annotation
    sine_starts = sines[0].astype(int)     # start coord of a sine song bout
    sine_ends = sines[1].astype(int)       # end coord of a sine song bout

    sine_cfs = sines[2]                    # carrier freq of that bout
    sine_cft = times[((sine_starts + sine_ends)/2).astype(int)] # plot CF at the center of the detected feature

    #unpack pulse annotation
    pulse_starts = pulses[1].astype(int)   # start coord of an individual pulse
    pulse_ends = pulses[2].astype(int)     # end coord of an individual pulse

    pulse_cfs = pulses[3]                  # carrier freq of that individual pulse
    pulse_cft = pulses[0]                  # time coord at center of that pulse
 
    #plot sines in red, pulses in blue
    for i in range(len(sine_starts)):
        if abs(times[sine_starts[i]]-times[sine_ends[i]])<1:
            sine_line = axs.plot(times[sine_starts[i]:sine_ends[i]],
                                    raw[sine_starts[i]:sine_ends[i],0],
                                    color='coral')

    for i in range(len(pulse_starts)):
        pulse_line = axs.plot(times[pulse_starts[i]:pulse_ends[i]],
                                raw[pulse_starts[i]:pulse_ends[i],0],
                                color='cadetblue')

    #legend
    axs.scatter([], [], c="coral", label="sine song", s=20)  # make an empty scatter plot to generate the legend from
    axs.scatter([], [], c="cadetblue", label="pulse song", s=20)
    #axs[0].legend(loc="best")
    #axs[0].set_title(file)
    axs.set_title(genotype)
    axs.set_ylim(ylim)

    axs2 = axs.twinx()

    # third y axis: LED stimulus
#     axs2.plot(times, raw[:,1], color='black', label='LED stimulation', alpha=.1)
#     axs2.set_ylabel('LED stim (V)')
#     axs2.set_xlabel('Time (sec)')
#     axs2.set_ylim(0,10)
    

    fig.legend(loc='center left', bbox_to_anchor=(1, 0.5))

    plt.tight_layout()
    plt.xlim(xlim)
    plt.show()
    
```

```python
def get_data_to_array(file_list, genotype, freq, fly):
    data = []
    for rep_i, file in enumerate(file_list):

        mat_tmp = loadmat(file)
        raw, times, sines, pulses = mat_tmp['dataout'], mat_tmp['times'],\
                        mat_tmp['sines'], mat_tmp['pulses']
        times = times.squeeze()

        good_sine_ids = np.abs(times[sines[0].astype(int)]-times[sines[1].astype(int)])<1 #not too long

        n_good_sine = sum(good_sine_ids)
        n_pulse = len(pulses[1])

        sine_freq = sines[2][good_sine_ids]
        sine_len = sines[1][good_sine_ids] - sines[0][good_sine_ids]

        pulse_freq = pulses[3]
        pulse_len = pulses[2] - pulses[1]   

        IPI = [b-e for b,e in np.array([times[pulses[1][1:]], times[pulses[2][:-1]]]).T]
        IPI.append(np.nan)
        mean_pulse_time = times[((pulses[1] + pulses[2])/2).astype(int)]
        start_pulse_time = times[(pulses[1]).astype(int)]

        N_sine_bouts = len(good_sine_ids)
        N_pulse_bouts = len(pulses[3])

        data.append(['N_pulse_bouts', genotype, freq, fly, file, 0, N_pulse_bouts])
        data.append(['N_sine_bouts', genotype, freq, fly, file, 0, N_sine_bouts])
        data.append(['N_bouts', genotype, freq, fly, file, 0, N_sine_bouts+N_pulse_bouts])


        for pulse_i in range(n_pulse):

            data.append(['pulse_freq', genotype, freq, fly, file, pulse_i, pulse_freq[pulse_i]])
            data.append(['pulse_len', genotype, freq, fly, file, pulse_i, pulse_len[pulse_i]])
            data.append(['IPI', genotype, freq, fly, file, pulse_i, IPI[pulse_i]])
            data.append(['mean_pulse_time', genotype, freq, fly, file, pulse_i, mean_pulse_time[pulse_i]])
            data.append(['start_pulse_time', genotype, freq, fly, file, pulse_i, start_pulse_time[pulse_i]])

            start_t_index = pulses[1][pulse_i].astype(int)
            mid_t_index = pulses[0][pulse_i].astype(int)
            end_t_index = pulses[2][pulse_i].astype(int)
            
            part1, part2 = np.array_split(raw[:,0][start_t_index:end_t_index], 2)
            
            if len(part1)>0:
                pulse_sym_score = np.dot(part1, np.flip(part2))

                data.append(['pulse_sym_score', genotype, freq, fly, file, pulse_i, pulse_sym_score])


        for sine_i in range(n_good_sine):

            data.append(['sine_freq', genotype, freq, fly, file, sine_i, sine_freq[sine_i]])
            data.append(['sine_len', genotype, freq, fly, file, sine_i, sine_len[sine_i]])
    
    return data
```

```python
all_data = []
cols = ['metric','genotype', 'freq', 'fly', 'file', 'songs', 'val']

for genotype in ['vpr6','pip10', 'p1']:
        
    for freq  in [0.001,0.01,0.1,1]:
        for fly in [1,2,4,6,8]:
            
            file_list = genotypes_all.get(genotype).get(freq).get(fly)

            if file_list is not None:
                all_data.extend(get_data_to_array(file_list, genotype, freq, fly))
                



file_list = genotypes_all['wt']
all_data.extend(get_data_to_array(file_list, 'wt' ,'999' ,1))
```
