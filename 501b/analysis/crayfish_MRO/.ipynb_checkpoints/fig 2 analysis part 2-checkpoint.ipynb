```python
greed_file_list = ['21929020', '21929013', '21929021', 
                       '21929019','21929010','21929015',
                      '21929022','21929014','21929017']


greed_distance = [5,10,15,
                  5,10,15,
                  5,10,15]
greed_speed = [1,1,1,
               3,3,3,
               6,6,6]

# greed_windows = [[32,39], [20.5,24], [55,57],
#                 [11.8,20], [7,9], [7.5,9],
#                 [10,17], [17,20], [10,13.5]]
```

```python
def get_pulls(data, pull_dis, t_dis_thresh = 500):
    '''
    return subarary including ids of each pull 
    '''
    # pulled defined by 2/3 of max pull
    pulled_ids = np.where(data['ch2'] > (.9)*pull_dis)[0]
    # split is the split of each pull, defined by a temporal dif thresh
    split_ids = np.where(np.diff(pulled_ids)>t_dis_thresh)[0]
    
    pulls = np.split(pulled_ids, split_ids)    
    return  pulls

def plot_pulls(data, pulls):
    plt.plot(data['times'], data['ch2'])
    for pull in pulls:
        plt.scatter(data['times'][pull], data['ch2'][pull])
    plt.show()   
        
#     plt.scatter(data['times'][pulled_ids[split_ids]], [6]*split_id.shape[0], color='yellow')
#     plt.scatter(data['times'][pulled_ids[split_ids+1]], 6, color='yellow')
```

```python
all_file_names_and_pull_id=[]
all_d_params=[]
all_s_params=[]
all_fit_param=[]
all_curve_data=[]
all_fit_r2=[]
params = {'height':65,'distance':25}
for file, d_param, s_param, in zip(greed_file_list, greed_distance, greed_speed):
    data_dir = '../../data/2_CrayfishMRO/better_data/'
    data = extract_paired(data_dir+file+'.abf')
    pulls = get_pulls(data, d_param)

    for pull_id, pull in enumerate(pulls):

        peaks_ids,_ = find_peaks(data['ch1'][pull], 
                         height=params['height'], 
                         distance=params['distance'])
        
        
        # get firing freq & timestamps
        ifr_val = get_ifr(data['times'][peaks_ids])
        ifr_times = np.convolve(data['times'][peaks_ids], np.ones(2), 'valid') / 2

        fit = fit_decay_curve(peaks_ids, data['times'][pull], True, 10)
        if fit:
            fit_param, curve_data, perr = fit
            
            #calc r2
#             residuals = curve_data[1] - decay_func(curve_data[0],*fit_param)
#             ss_res = np.sum(residuals**2)
#             ss_tot = np.sum((curve_data[1]-np.mean(curve_data[1]))**2)
#             r2 = 1 - (ss_res / ss_tot)
            r2 = r2_score(curve_data[1], decay_func(curve_data[0],*fit_param))


            all_file_names_and_pull_id.append([file, pull_id])
            all_d_params.append(d_param)
            all_s_params.append(s_param)
            all_fit_param.append(fit_param)
            all_curve_data.append(curve_data)
            all_fit_r2.append(r2)
        
    


```

```python
for d_param, s_param, curve_data, fit_data, r2 in zip(all_d_params, all_s_params, all_curve_data, all_fit_param, all_fit_r2):
    print(d_param, s_param, fit_data[1])
    print(r2)
    plt.scatter(curve_data[0],curve_data[1])
    plt.plot(curve_data[0], decay_func(curve_data[0],*fit_data), color='coral')
    plt.show()
    
```

```python
all_tau_values = np.zeros([3,3,3])
all_r0_values = np.zeros([3,3,3])
all_rt_values = np.zeros([3,3,3])
all_r2_values = np.zeros([3,3,3])

for s_id, speed in enumerate([1,3,6]):
    for d_id, distance in enumerate([5,10,15]):
        val_ids = np.where((np.array(all_d_params)==distance) & (np.array(all_s_params)==speed))
        
        tau_vals = np.array(all_fit_param)[val_ids[0]][:,1]
        r0_vals = np.array(all_fit_param)[val_ids[0]][:,0]
        rt_vals = np.array(all_fit_param)[val_ids[0]][:,2]
        r2_vals = np.array(all_fit_r2)[val_ids[0]]

        #data_len = all_curve_data[val_ids[0]]
        for p_id, _ in enumerate(tau_vals):
            data_len = np.array(all_curve_data)[val_ids[0][p_id]][1].shape[0]
            if data_len>30:
                if tau_vals[p_id]>5:
                    tau_val = np.nan
                else:
                    tau_val=tau_vals[p_id]
                
                all_tau_values[s_id][d_id][p_id] =  tau_val
                all_r0_values[s_id][d_id][p_id] = r0_vals[p_id]
                all_rt_values[s_id][d_id][p_id] = rt_vals[p_id]
                all_r2_values[s_id][d_id][p_id] = r2_vals[p_id]
              
    #         if tau_vals.shape[0]==3:
    #             all_tau_values[s_id][d_id] = tau_vals
#         else:
#             all_tau_values[s_id][d_id] = np.hstack([tau_vals,np.mean(tau_vals)])

all_tau_values[all_tau_values == 0] = np.nan
all_r0_values[all_r0_values == 0] = np.nan
all_rt_values[all_rt_values == 0] = np.nan
all_r2_values[all_r2_values == 0] = np.nan

```
